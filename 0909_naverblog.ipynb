```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

import pandas
import numpy

from bs4 import BeautifulSoup
import lxml

import time
```

```python
path = '/chromedriver'
driver = webdriver.Chrome(path)
driver.implicitly_wait(3)

start_url = 'http://www.naver.com'
driver.get(start_url)
```

```python
#Keys.RETURN 했는데 안먹히면 Keys.ENTER로
#send_keys에서 click포함해서 따로 쓸 필요가 없음
search_box = driver.find_element_by_xpath('//*[@id="query"]')
search_box.send_keys('코로나백신')
search_box.send_keys(Keys.ENTER)
```

```python
#view 클릭하기
driver.find_element_by_xpath('//*[@id="lnb"]/div[1]/div/ul/li[3]/a').click()
```

```python
#볼로그 클릭하기
driver.find_element_by_xpath('//*[@id="snb"]/div[1]/div/div[1]/a[2]').click()
```

```python
#무한스크롤
SCROLL_PAUSE_TIME = 0.5

#body로 지정
body = driver.find_element_by_tag_name('body')

while True:
    last_height = driver.execute_script('return document.documentElement.scrollHeight')
    #현재 화면의 길이를 return받아서 last_height로 저장

    for i in range(10):
        body.send_keys(Keys.END) #body 분문에 END를 넣어주는거 (스크롤 내려줌)

        time.sleep(SCROLL_PAUSE_TIME) #데이터로딩시간을 주기워해서 스크롤 후 휴식
    
    new_height = driver.execute_script('return document.documentElement.scrollHeight')
    if new_height == last_height:
        break
```

```python
#Beautiful Soup
page = driver.page_source
soup = BeautifulSoup(page, 'html.parser')
driver.close()
```

```python
len(soup.text)
```

```python
blogs = soup.find_all('a', attrs = {'class': 'api_txt_lines total_tit'})
blog_list = []
for i in blogs:
    blog = i.get_text()
    blog_list.append(blog)
print(blog_list)
```

```python
IDs = soup.find_all('span', attrs = {'class': 'elss etc_dsc_inner'})
ID_list = []
for i in IDs:
    ID = i.get_text()
    ID_list.append(ID)
print(ID_list)
```

```python
contents = soup.find_all('div', attrs = {'class': 'api_txt_lines dsc_txt'})
content_list = []
for i in contents:
    content = i.get_text()
    content_list.append(content)
print(content_list)
```

```python
import pandas as pd
wedtoon_df = pd.DataFrame({'제목': blog_list, '작성자': ID_list, '내용': content_list})
wedtoon_df.head()
```
